```python
user_prompt = PromptTemplate(
    input_variables=["question"],
    template="{question}"
)

print(user_prompt.format(question="What is AI?"))
```

```python
from langchain_openai import ChatOpenAI
from langchain_classic.chains import LLMChain

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)

chain = LLMChain(
    llm=llm,
    prompt=user_prompt
)

response = chain.invoke({"question": "What is AIR?"})
print(response["text"])
```
